### Task 11
Create a new column titled `CS5000 Grade` in the score DataFrame to calcualte the letter grade for each student in course CS 5000 using the following criteria
+ `A: score >= 90`
+ `B: 80 <= score < 90`
+ `C: 70 <= score < 80`
+ `D: 60 <= score < 70`
+ `F: score < 60`

```python
grades=[]
for i in df2['CS 5000']:
    if(i>=90):
        grades.append('A')
    elif (80 <= i < 90):
        grades.append('B')
    elif(70 <= i < 80):
        grades.append('C')
    elif(60 <= i < 70):
        grades.append('D')
    else:
        grades.append('F')
df2['CS5000 Grade']=grades
df2
```
